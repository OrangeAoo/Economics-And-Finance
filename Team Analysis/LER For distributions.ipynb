```python
import pandas as pd;
import seaborn as sns;
import numpy as np;
import matplotlib.pyplot as plt;
from scipy.special import gammainc, gamma;
from scipy.stats import norm;
```

```python

# LER for an exponential with mean = 10
def ler(d):
    return 1 - np.exp(-d / 10)

d = np.arange(0, 35.1, 0.1)
ler_values = ler(d)

x = pd.DataFrame({'d': d, 'ler': ler_values})

sns.lineplot(data=x, x='d', y='ler')

plt.title('LER for the Exponential w/ Mean = 10')
plt.xlabel('deductible')

plt.show;
```

```python
##Gamma
def ler(d):
    alpha = 2
    theta = 10
    ex = alpha * theta
    exd = alpha * theta * gammainc(alpha + 1, d / theta) + d * (1 - gammainc(alpha, d / theta))
    return exd/ex

d = np.arange(0, 35.1, 0.1)
ler_values = ler(d)

x = pd.DataFrame({'d': d, 'ler': ler_values})

sns.lineplot(data=x, x='d', y='ler')

plt.title('LER for the Gamma w/ alpha = 2 and theta = 10')
plt.xlabel('deductible')

plt.show;
```

```python
#Pareto alpha != 1
def ler(d):
    alpha = 2
    theta = 10
    ex = theta/(alpha - 1)
    exd = (theta/(alpha - 1))*(1 - (theta/(theta+d))**(alpha-1))
    return exd/ex

d = np.arange(0, 35.1, 0.1)
ler_values = ler(d)

x = pd.DataFrame({'d': d, 'ler': ler_values})

sns.lineplot(data=x, x='d', y='ler')

plt.title('LER for the Pareto w/ alpha = 2 and theta = 10')
plt.xlabel('deductible')

plt.show;
```

```python
##Lognormal
def ler(d):
    mu = 0
    sigma = 1
    ex = np.exp(mu+((sigma)**2)/2)
    exd = np.exp(mu+((sigma)**2)/2)*norm.cdf((np.log(d) - mu - sigma**2) / sigma)+d * (1 - norm.cdf((np.log(d) - mu) / sigma))
    return exd/ex


d = np.arange(0, 35.1, 0.1)
ler_values = ler(d)

x = pd.DataFrame({'d': d, 'ler': ler_values})

sns.lineplot(data=x, x='d', y='ler')

plt.title('LER for the Lognormal w/ Mean = 0 and sigma = 1')
plt.xlabel('deductible')

plt.show;
```
