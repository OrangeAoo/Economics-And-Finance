```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
```

## Gamma Distribution

>One-parameter distribution suitable for small claims


```python
# Setting up the plotting area
fig, axes = plt.subplots(1, 2, figsize = (12, 6))
plt.subplots_adjust(wspace = 0.3)

# Varying Scale Gamma Densities
scaleparam = np.arange(100, 251, 50)
shapeparam = np.arange(2, 6)
x = np.arange(0, 1001, 1)

axes[0].set_ylabel("Gamma Density")

for k, scale in enumerate(scaleparam, start = 1):
    fgamma = gamma.pdf(x, a = 2, scale = scale)
    axes[0].plot(x, fgamma, label = f'scale = {scale}', color = f'C{k}')

axes[0].legend(loc="upper right")
axes[0].set_title('Shape = 2')

# Varying Shape Gamma Densities
for k, shape in enumerate(shapeparam, start = 1):
    fgamma = gamma.pdf(x, a = shape, scale = 100)
    axes[1].plot(x, fgamma, label = f'shape = {shape}', color = f'C{k}')


plt.figtext(0.5, 1, "Gamma Densities", ha = 'center', fontsize = 14)
axes[1].legend(loc = "upper right")
axes[1].set_title('Scale = 100')

plt.show()
```

## Loss Elimination Ratio
Ratio of the decrease in the expected payment (with deductible) to the expected payment without deductible.

```python
import pandas as pd
import seaborn as sns

# LER for an exponential with mean = 10
def ler(d):
  return 1 - np.exp(-d / 10)

d = np.arange(0, 35.1, 0.1)
ler_values = ler(d)

x = pd.DataFrame({'d': d, 'ler': ler_values})

sns.lineplot(data=x, x='d', y='ler')

plt.title('LER for the Exponential w/ Mean = 10')
plt.xlabel('deductible')

plt.show;
```
